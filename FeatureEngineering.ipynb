```python
import pandas as pd
import numpy as np 
```

```python
train = pd.read_csv('train.csv')
macro = pd.read_csv('macro.csv')
```

```python
nanmacro = np.array(macro.isnull().sum())
trimmed_macro = macro.iloc[:,nanmacro < 10]
print(trimmed_macro.shape)
trimmed_macro.head()
```

```python
nantrain = np.array(train.isnull().sum())
trimmed_train = train.iloc[:,nantrain == 0]
print(trimmed_train.shape)
trimmed_train.head()
```

```python
merged = trimmed_train.merge(trimmed_macro, on='timestamp', how='left')
print(merged.shape)
merged.head()
```
